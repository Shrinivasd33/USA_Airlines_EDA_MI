#### <font color='blue'>Clean up data for EDA </font> 

```python
data_raw.drop_duplicates(keep=False, inplace=True)
data_raw=data_raw.dropna(subset=['ActualElapsedTime','CRSElapsedTime'])
data_raw = data_raw.loc[data_raw["AirTime"] >= 10 ]
data_raw.shape
```

#### <font color='blue'>Write a function to get name weekdays from weekday number</font> 

```python
week_day =  {0:'Mon',1:'Tue',2:'Wed',3:'Thru',4:'Fri',5:'Sat',6:'Sun'}

def getweekname(n):
    for key, value in week_day.items():
        if key == n:
            return value
        
data_raw["WeekDay"] = data_raw["DayOfWeek"].apply(getweekname)

data_raw.head(2)
```

#### <font color='blue'>Get the Carrier code and merge with Raw data</font> 

```python
data_raw = data_raw.merge(carriers, left_on='UniqueCarrier', right_on='Code', how='inner')
```

## <font color='blue'>Seperate Cancelled and Non_Cancelled in Seperate dataframes </font> 

```python
data_raw.groupby(by='Cancelled').size()
```

```python
plane_canc = data_raw[data_raw['Cancelled'] == 1]
plane_canc.shape
```

```python
plane_oper = data_raw[data_raw['Cancelled'] == 0]
plane_oper.shape
```

```python
plane_canc_df = plane_canc.groupby(['Description']).size().sort_values(ascending = False).reset_index().rename(columns={0: 'CANCcount'})
plane_canc_df.head(8)
```

```python
plane_oper_df = plane_oper.groupby(['Description']).size().sort_values(ascending = False).reset_index().rename(columns={0: 'OperCount'})
plane_oper_df.head(8)
```

#### <font color='blue'> Planes Cancelled ratio

```python
plane_oper_ratio = plane_oper_df.merge(plane_canc_df, on='Description', how = 'left')
plane_oper_ratio['CANCcount'] = plane_oper_ratio['CANCcount'].fillna(0)
plane_oper_ratio.round(2).sort_values('CANCcount', ascending = False)
```

```python
plane_oper_ratio['CNCL_Ratio'] = (plane_oper_ratio['CANCcount'] / plane_oper_ratio['OperCount'])*100
plane_oper_ratio.round(2).sort_values('CNCL_Ratio', ascending = False)
```

#### <font color='blue'> Get Cancel reasons

```python
cancle_code = {'A':'CarrierIssue', 'B':'BadWeather', 'C': 'NationalAviationSys', 'D': 'SecurityDelayed'}

def getcanceldetails(n):
    for keys, values in cancle_code.items():
        if n == keys:
            return values
        
plane_canc["CancellationReasons"] = plane_canc["CancellationCode"].apply(getcanceldetails)
plane_canc.head(4)
```

```python
plane_canc.groupby(['CancellationReasons','Description']).size().reset_index().rename(columns={0:'Count'}).sort_values(['Count'], ascending = False)
```

```python
plane_canc_route = plane_canc.groupby(['RouteKey','CancellationReasons','Description']).size().reset_index().rename(columns={0:'Count'}).sort_values('Count',ascending = False)
plane_canc_route[(plane_canc_route['Count'] > 60)]
```
